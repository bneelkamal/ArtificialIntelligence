Simple Reflex Agent

```python
import numpy as np
import time



class SimpleReflexAgent :  
    def __init__(self, environment):
        self.location = (0,0)
        self.environment = environment

    def perceive(self): #perceive the environment
        return self.environment[self.location[0], self.location[1]] == "DIRTY" # if dirty, return true

    def act(self): #perform the action and move
        if self.perceive():
            print(f"Location {self.location} is DIRTY. Cleaning ...")
            self.environment[self.location] = "CLEAN"
        else:
            print(f"Location {self.location} is CLEAN. Moving ...")
            self.move()

    def move(self): #decide the next move
        rows, cols = self.environment.shape
        row, col = self.location
        if col+1 < cols:
            self.location = (row,col+1)
        elif row+1 < rows:
            self.location = (row+1,0)
        else:
            print("no more moves.cleaning complete!")

        

environment = np.array([
    ["DIRTY","CLEAN","DIRTY"],
    ["CLEAN","DIRTY","CLEAN"],
    ["DIRTY","DIRTY","CLEAN"]
])

print(f"initial environment : \n{environment}\n\n ")

agent = SimpleReflexAgent(environment)

while True:
    if np.all(agent.environment == "CLEAN"):
        print("all locations are clean. stopping.")
        break

    agent.act()
    print(f"current environment : \n{environment}\n\n ")


```
